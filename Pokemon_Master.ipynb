```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
df = pd.read_csv('pokemon.csv')
```

```python
def select_columns(dataf):
    return dataf.loc[:,['pokedex_number','name','type1','weight_kg','height_m','hp','attack','defense']]
pokemon = df.pipe(select_columns)
```

```python
pd.set_option('max_rows', 20)
```
